```python
import codecs
import re
# API_name = 'java.util.Collections'


file_API = 'tutorial_col_API.txt' # API file name
file = codecs.open("javadoc_col_short", 'w','utf-8') #Output file
input_file = 'docs.txt' #input file

#Reading API name
with open(file_API,'r') as f1:
    for line1 in f1:
        line1 = line1.replace('\r', '').replace('\n', '')
        print (line1)
        API_name = line1
        pattern = re.compile("^"+API_name+"[^A-Za-z0-9_]|"+API_name+"$")
        
        # Read input file
        with open(input_file,'r') as f2:
            count = 0
            combine_str=""
            for line2 in f2:  
                word_split = line2.split(':',1)
                check = pattern.match(word_split[0])
                if check is not None:
                    # Combine method

                    if (count is 0):
                        paragraph = word_split[1].replace('\r', '').replace('\n', '')
                        file.write(word_split[0] +": "+ paragraph+'\n')
                        combine_str =word_split[0] +": "+ paragraph+' '
                    else:
                        paragraph = word_split[1].replace('\r', '').replace('\n', '')
                        print (word_split[0])
                        file.write(combine_str + word_split[0] +": "+ paragraph+'\n')

#                     paragraph = word_split[1].replace('\r', '').replace('\n', '')
#                     print (word_split[0])
#                     file.write(API_name + ": "+combine_str + word_split[0] +": "+ paragraph)
#                     file.write('\n') // Concatinate all paragraph
                    count += 1
#         file.write('\n')
        print (API_name+": "+str(count))
```
