```python
#FINAL VERSION
import glob
import os
import codecs
from pathlib import Path
import pandas as pd
from bs4 import BeautifulSoup

# Read all csv file in the path
# Input: the path
# output: content without html + oracle

path = 'E:/UTD/Research/fragmentation/McGill/'
# fileDir = os.path.dirname(os.path.realpath('__file__'))
# path = os.path.join(fileDir, '../McGill/')

# path = '/McGill/'
for file_name in glob.glob(os.path.join(path, '*.csv')):
    print ("READING FILE: " + file_name)

    # Format name of output file    
    index_of_slash = file_name.index('\\') #eliminate '\'
    out_file = file_name[index_of_slash+1:]
    index_of_dot = out_file.index('.') #eliminate "."
    out_file = out_file[:index_of_dot]
    my_file = Path(out_file)
    
    # Open output file
    file = codecs.open(out_file, encoding='utf-8', mode='w') 
    oracle_file = codecs.open(out_file+"_oracle", encoding='utf-8', mode='w') 
    
    # Load spreadsheet
    xl = pd.read_csv(file_name,header=None)
    old_content =""
    count = 0
    for i in range(0,len(xl)):   
        html_doc = xl[3][i] #Read desired content
        api_name = xl[5][i]
        clf_result = xl[7][i]
        new_content = xl[2][i] 
        soup = BeautifulSoup(html_doc, 'html.parser')
        if (old_content != new_content): #check duplication
            s = soup.get_text()
            s = " ".join(s.split()) # this removes "all whitespace characters(space, tab, newline, return, formfeed)", into 1 single space
#             print (s)
            file.write(s + "\n")
            if old_content != "":
                oracle_file.write("\n")
                print("\n")
            count += 1
#         else:
        oracle_file.write(api_name + " " + str(clf_result) + " ")
        print (api_name + " " + str(clf_result) + " ",end="")
        old_content = new_content
    print(count)
    file.close()
    oracle_file.close()
```
